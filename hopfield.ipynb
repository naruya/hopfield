# 課題A Hopfield型ニューラルネットワーク

類似度 $D$ は、ハミング距離 $H$ 、画像の全要素数 $S$ を用いて、  
$ \quad D = 1-H/S $  
とした。

各性能は、100回ずつ実行したときの平均を用いた。

### 準備

```python
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy

np.random.seed(2)
```

```python
pattern = np.load("pattern.npy")

plt.figure(figsize=(12,2))
for i,pat in enumerate(pattern):
    plt.subplot(1,6,i+1)
    plt.imshow(pat)
plt.show()
```

```python
def simularity(A, B):
    return 1-np.sum(np.abs(A-B))/(2*A.shape[0]*A.shape[1]) # abs((-1)-(1))==2
```

```python
class HopfieldNetwork:
    def __init__(self, shape):
        self.W = np.zeros(shape**2)

    def set_img(self, img):
        self.X = np.copy(img).reshape(1,len(self.W))
    
    def get_img(self):
        return self.X
    
    def train(self, pattern):
        n_pattern = len(pattern)
        pattern = pattern.reshape((n_pattern, 1, -1))
        for pat in pattern:
            dW = pat.T * pat
            self.W += dW
            
    def recall(self):
        index = np.random.randint(len(self.W))
        self.X[0,index] = np.where(self.X.dot(self.W[:,index])>0, 1, -1)
        
    def energy(self):
        return -self.X.dot(self.W).dot(self.X.T)[0,0]
```

```python
def noise(image, rate):
    mask = np.random.uniform(0,1,image.shape)
    mask = np.where(mask>(1-rate), -1, 1)
    return mask * image 
```

### 想起の様子の可視化

```python
def hopfield_show(n_pattern, noise_rate):
    print(str(n_pattern)+"種類, noise_rate == {:.1f}".format(noise_rate))

    hp = HopfieldNetwork(np.array([5,5]))

    # 1パターンのみで学習
    hp.train(pattern[:n_pattern])

    epochs = 200

    hist = []
    image = pattern[0]

    # plot用
    original = deepcopy(image)

    noise_image = noise(image, noise_rate)
    hp.set_img(noise_image)

    # plot用
    before = deepcopy(hp.get_img().reshape(image.shape))

    plt.figure(figsize=(10,1))

    for epoch in range(epochs):

        hist.append(hp.energy())
        hp.recall()

        if epoch % 20 == 0:
            img = hp.get_img().reshape(image.shape)
            plt.subplot(1,10,(epoch//20)+1); plt.title("#"+str(epoch)); plt.axis('off'); plt.imshow(img)

    plt.show()

    # plot用
    after = deepcopy(hp.get_img().reshape(image.shape))

    plt.figure(figsize=(3,1))
    plt.subplot(1,3,1); plt.title("original"); plt.axis('off'); plt.imshow(image)
    plt.subplot(1,3,2); plt.title("before"); plt.axis('off'); plt.imshow(before)
    plt.subplot(1,3,3); plt.title("after"); plt.axis('off'); plt.imshow(after)
    plt.show()

    plt.figure(figsize=(3,2))
    plt.ylabel("energy")
    plt.xlabel("epoch")
    plt.plot(hist)
    plt.show()

hopfield_show(1, 0.2)
```

```python
def hopfield_sim_acc(n_pattern, noise_rate):
    print(str(n_pattern)+"種類, noise_rate == {:.1f}".format(noise_rate))
    simularities = []

    for _ in range(100):
        hp = HopfieldNetwork(np.array([5,5]))

        # nパターンのみで学習
        hp.train(pattern[:n_pattern])
        
        epochs = 200

        hist = []
        image = pattern[0]

        original = deepcopy(image)

        noise_image = noise(image, noise_rate)
        hp.set_img(noise_image)

        before = deepcopy(hp.get_img().reshape(image.shape))

        for epoch in range(epochs):
            hist.append(hp.energy())
            hp.recall()

        after = deepcopy(hp.get_img().reshape(image.shape))

        simularities.append(simularity(original, after))
    
    sim = np.mean(simularities)*100
    acc = np.mean(np.where(np.array(simularities)==1, 1., 0.))*100
    print("類似度:", sim, "%")
    print("正答率:", acc, "%")
    print()
    return sim, acc
```

# 課題A-1
- 1~6種類の画像を覚えさせ、元画像にノイズを加えた画像を初期値とし実行

```python
value = (hopfield_sim_acc(1, 0.2), 
         hopfield_sim_acc(2, 0.2), 
         hopfield_sim_acc(3, 0.2), 
         hopfield_sim_acc(4, 0.2), 
         hopfield_sim_acc(5, 0.2), 
         hopfield_sim_acc(6, 0.2))
```

```python
value = np.array(value)
plt.plot(value[:,0], label="similarity")
plt.plot(value[:,1], label="accuracy")
plt.legend()
plt.show()
```

# 課題A-2
- 画像が2種類と4種類の場合について、ノイズを0~100にしたときの想起性能

```python
value2 = []
value4 = []

for r in range(5):
    hopfield_show(2, 0.1+0.2*r)
    value2.append(hopfield_sim_acc(2, 0.1+0.2*r))
    print("-"*80)
for r in range(5):
    hopfield_show(4, 0.1+0.2*r)
    value4.append(hopfield_sim_acc(4, 0.1+0.2*r))
    print("-"*80)
```

```python
value2 = np.array(value2)
value4 = np.array(value4)
plt.title("2 pattern vs 4 pattern")
plt.xlabel("noise_rate")
plt.ylabel("similarity / accuracy")
x = [0.1, 0.3, 0.5, 0.7, 0.9]
plt.plot(x, value2[:,0], label="2 similarity", c="b")
plt.plot(x, value2[:,1], label="2 accuracy", linestyle=":", c="b")
plt.plot(x, value4[:,0], label="4 similarity", c="r")
plt.plot(x, value4[:,1], label="4 accuracy", linestyle=":", c="r")
plt.legend()
plt.show()
```

# 課題A-3
- 自分の疑問: ノイズを50%以上にした時、元画像とは逆の画像の想起率が上がっているか

```python
def hopfield_inv_sim_acc(n_pattern, noise_rate):
    print(str(n_pattern)+"種類, noise_rate == {:.1f}".format(noise_rate))
    simularities = []

    for _ in range(100):
        hp = HopfieldNetwork(np.array([5,5]))

        # nパターンのみで学習
        hp.train(pattern[:n_pattern])
        
        epochs = 200

        hist = []
        image = pattern[0]

        original = deepcopy(image)

        noise_image = noise(image, noise_rate)
        hp.set_img(noise_image)

        before = deepcopy(hp.get_img().reshape(image.shape))

        for epoch in range(epochs):
            hist.append(hp.energy())
            hp.recall()

        after = deepcopy(hp.get_img().reshape(image.shape))

        simularities.append(simularity(-original, after))
    
    sim = np.mean(simularities)*100
    acc = np.mean(np.where(np.array(simularities)==1, 1., 0.))*100
    print("類似度:", sim, "%")
    print("正答率:", acc, "%")
    print()
    return sim, acc
```

```python
value_org = []
value_inv = []

print("---- original ----\n")
for r in range(5):
    value_org.append(hopfield_sim_acc(2, 0.1+0.2*r))

print("---- inverse ----\n")
for r in range(5):
    value_inv.append(hopfield_inv_sim_acc(2, 0.1+0.2*r))
```

```python
value_org = np.array(value_org)
value_inv = np.array(value_inv)
plt.xlabel("noise_rate")
plt.ylabel("similarity / accuracy")
x = [0.1, 0.3, 0.5, 0.7, 0.9]
plt.plot(x, value_org[:,0], label="org similarity", c="b")
plt.plot(x, value_org[:,1], label="org accuracy", linestyle=":", c="b")
plt.plot(x, value_inv[:,0], label="inv similarity", c="r")
plt.plot(x, value_inv[:,1], label="inv accuracy", linestyle=":", c="r")
plt.legend()
plt.show()
```

# 考察

A-1
- 覚えさせるパターンを増やすとだいたいaccuracyとsimularityは下がったが、6パターン覚えさせたときと比べると、4,5パターン覚えさせたときに極端にaccuracyが下がっていた。これは、覚えたパターンの中によく似たパターンのペア（ここでは『C』と『X』）があってかつ覚えたパターンの数が少ない場合、よく似たパターンの影響を受けて『C』と『X』の中間的な画像、具体的には

  

  がエネルギー的に安定な画像となってしまい想起に失敗してしまったためだと考える。また5枚目を追加したときも、『X』と『Y』が似ていたためより中間的な画像がより出やすくなってしまっていたようであった。6パターン目を追加すると、覚えた全部のパターン数に対して似たパターンのペアの割合が減ったため、想起が少し成功しやすくなったと考える。

A-2
- 2枚覚えさせたときより4枚覚えさせたときのほうが全体的にaccuracyが下がった。similarityはそこまで変わらなかった。

A-3
- ノイズを大きくしたときに元画像とは逆の画像の想起率が上がっていることが確認できた。これはHopfieldネットワークは相対的な情報、つまりノード間の関係性は評価されるものの絶対的な情報は評価されないため、真逆の画像に想起してもエネルギー的に安定になってしまうためである。
